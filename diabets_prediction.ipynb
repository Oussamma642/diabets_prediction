### Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score 
```

### Data Collection and Analysis

```python
# loading the diabets dataset to a pandas dataframe
diabets_dataset = pd.read_csv('../diabets_prediction/diabetes.csv')
```

```python
# Printing first five lines
diabets_dataset.head()
```

```python
# Number of rows and columns
diabets_dataset.shape
```

```python
# Getting the statistical measures of the dataset
diabets_dataset.describe()
```

```python
diabets_dataset['Outcome'].value_counts()
```

```python
# Mean of diabets and non diabets 
diabets_dataset.groupby('Outcome').mean()
```

```python
# Seperating the data and labels
X = diabets_dataset.drop(columns='Outcome', axis=1) # axis = 1 to indicate that we are dropping a column
Y = diabets_dataset['Outcome']
```

### Data standarization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.fit_transform(X)
```

```python
X = standardized_data
Y = diabets_dataset['Outcome']
```

### Train Test split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

### Training The Model

```python
classifier = svm.SVC(kernel='linear')
```

```python
# Training the support vector machine classifier 
classifier.fit(X_train, Y_train)
```

### Evaluate the model

##### Accuracy Score on training data

```python
# Accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(f"The accuracy score of the training data: {training_data_accuracy}")
```

```python
# Accuracy score on the training data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(f"The accuracy score of the test data: {test_data_accuracy}")
```
